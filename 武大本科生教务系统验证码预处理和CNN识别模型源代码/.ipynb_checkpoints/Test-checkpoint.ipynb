```python
import cv2
import os
import pickle
import numpy as np
from PIL import Image
from helpers import resize_to_fit  
from JpgPretreat import twoValue,clearNoise,saveImage,cfs,saveSmall,convert
from tensorflow.keras.models import load_model

IMAGEFORTEST_FOLDER = "test"

a = input("是否区分大小写(y or n): ")
while a != 'y' and a != 'n':
    a = input("是否区分大小写(y or n): ")

if a == 'y':
    MODEL_LABELS_FILENAME = "model_labels_.dat"
    model = load_model('captcha_model_.hdf5')
else:    
    MODEL_LABELS_FILENAME = "model_labels.dat"
    model = load_model('captcha_model.hdf5')
    

with open (MODEL_LABELS_FILENAME, 'rb') as f:
    lb = pickle.load(f)

    
for file in os.listdir(IMAGEFORTEST_FOLDER):
    
    image = Image.open('./'+IMAGEFORTEST_FOLDER+'/'+file).convert("L")
    # 二值化
    twoValue(image, 100) 
    # 降噪
    clearNoise(image, 1, 1)
    image=saveImage(image.size)
    # 切割
    x_cuts,y_cuts = cfs(image)

    file = file.split('.')[0].lower() if a == 'n' else file.split('.')[0]
    
    print("Label".rjust(10), end = ": ")
    for c in file:
        print(c, end = ' ')
    print('\nPrediction', end = ": ")
    
    for j, item in enumerate(x_cuts):
        if j<4:
            box = (item[0], 0, item[1], 30)
            tmp = image.crop(box)
            tmp.save('./'+'test_1'+'/'+file+'_'+file[j]+'.png')
            img = cv2.imread('./'+'test_1'+'/'+file+'_'+file[j]+'.png') 
#            os.remove('./'+'test_1'+'/'+file+'_'+file[j]+'.png')
            img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            # Resize the letter so it fits in a 20x20 pixel box
            img = resize_to_fit(img, 30, 30) # 引入的函数

            # Add a third channel dimension to the image to make Keras happy
            img = np.expand_dims(img, axis=2)            

            nm_arr = img.reshape([1,30,30,1])
            nm_arr = np.array(nm_arr, dtype="float") / 255.0
            

            predicted_cat = model.predict(nm_arr)
            print(lb.inverse_transform(predicted_cat)[0],end=' ')

    print('\n')

print('end')
```
